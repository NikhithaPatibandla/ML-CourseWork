```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Upload the dataset
from google.colab import files
uploaded = files.upload()
```

```python
#Load the dataset
df = pd.read_csv('Day_7_sales_data.csv')
df
```

```python
print("Dataset Preview: ")
df.head() #prints the first 5 rows
```

```python
#Check for any missing values
print("\nChecking for missing values: ")
df.isnull().sum()
```

```python
#Give more than one independent variable and one dependent variable
X = df[['Profit','Quantity']]
y = df['Sales']
```

```python
#Split the dataset into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Apply the linear regression algorithm and train the dataset
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Print the Model intercept and coefficients
print("\nModel Coefficients: ")
print("Intercept: ",model.intercept_)
print("Coefficients: ",model.coef_)
```

```python
#Predicting the model from train data set
y_pred = model.predict(X_test)
# Calculate the metrics
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
#Print the evaluation metrics
print("\nModel Evaluation: ")
print("Mean Squared Error: ",mse)
print("R-squared: ",r2)
```

```python
#Plot the output
plt.scatter(y_test, y_pred,color='blue',alpha=0.5)
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.grid()
plt.show()
```
